```python
from tkinter import *
import math
PINK = "#e2979c"
RED = "#e7305b"
GREEN = "#9bdeac"
YELLOW = "#f7f5dd"
FONT_NAME = "Courier"
WORK_MIN = 25
SHORT_BREAK_MIN = 5
LONG_BREAK_MIN = 20
 TEXT = "âœ”"
reps = 0
timer = None
```

```python
# --------------------------TIMER RESET----------------------------- # 
def reset_timer():
    global reps
    window.after_cancel(timer)
    canvas.itemconfig(timer_text, text="00:00")
    title_label.config(text="Timer")
    check_mark.config(text="")
    reps = 0
```

```python
# --------------------------TIMER MECHANISM----------------------------- # 
def start_timer():
    global reps
    reps+=1
    
    work_sec = WORK_MIN * 60
    short_break = SHORT_BREAK_MIN * 60
    long_break = LONG_BREAK_MIN * 60
#     while True:       
#         if reps == 1 or reps == 4 or reps == 6:
#             count_down(short_break)
#             title_label.config(text="Break", fg=PINK, bg=YELLOW, font=(FONT_NAME, 50))
#             reps += 1
#         elif reps == 8:
#             count_down(long_break)
#             title_label.config(text="Break", fg=RED, bg=YELLOW, font=(FONT_NAME, 50))
#         else:    
#             count_down(work_sec)
#             title_label = Label(text="Work", fg=GREEN, bg=YELLOW, font=(FONT_NAME, 50))
#             reps +=1 
    if reps%8 == 0:
        count_down(long_break)
        title_label.config(text="Break", fg=RED, bg=YELLOW, font=(FONT_NAME, 50))
    elif reps%2 == 0:
        count_down(short_break)
        title_label.config(text="Break", fg=PINK, bg=YELLOW, font=(FONT_NAME, 50))
    else:
        count_down(work_sec)
        title_label.config(text="Work", fg=GREEN, bg=YELLOW, font=(FONT_NAME, 50))
    
```

```python
# ------------------------COUNTDOWN MECHANISM--------------------------- # 
def count_down(count):
    global reps
    count_min = math.floor(count / 60)
    count_sec = count % 60 
    if count_sec == 0 and count_min > 0:
        count_sec = "00"
    if int(count_sec) < 10 and count_min == 0:
        count_sec = f"0{count_sec}"
    canvas.itemconfig(timer_text, text=f"{count_min}:{count_sec}")
    if count > 0:
        global timer
        timer = window.after(1000, count_down, count-1)
    else:
        start_timer()
        mark = ""
        work_sessions = math.floor(reps/2)
        for _ in range(work_sessions):
            mark += TEXT
        check_mark.config(text=TEXT, fg=GREEN, bg=YELLOW)
```

```python
# ----------------------------UI SETUP------------------------------- #
window = Tk()
window.title("Tamatar")
window.config(padx=100, pady=50, bg=YELLOW)


title_label = Label(text="Timer", fg=GREEN, bg=YELLOW, font=(FONT_NAME, 50))
title_label.grid(column=1, row=0)

canvas = Canvas(width=200, height=224, bg=YELLOW, highlightthickness=0)
tomato_img = PhotoImage(file="tomato.png")
canvas.create_image(100, 112, image=tomato_img)
timer_text = canvas.create_text(100, 130, text="00:00", fill="light yellow", font=(FONT_NAME,35,"bold"))
canvas.grid(column=1, row=1)

button_border_start = Frame(window, highlightbackground = "black", 
                         highlightthickness = 0.5, bd=0)
button_border_start.grid(column=0, row=2)

button_border_reset = Frame(window, highlightbackground = "black", 
                         highlightthickness = 0.5, bd=0)
button_border_reset.grid(column=2, row=2)


def changeOnHover(button, colorOnHover, colorOnLeave):
    button.bind("<Enter>", func=lambda e: button.config(background=colorOnHover, fg=RED)) 
    button.bind("<Leave>", func=lambda e: button.config(background=colorOnLeave, fg="black"))

start_btn = Button(button_border_start, text="Start", bg=YELLOW, command=start_timer)
start_btn.grid(column=0, row=2)
changeOnHover(start_btn, GREEN, YELLOW)

reset_btn = Button(button_border_reset, text="Reset", bg=YELLOW, command=reset_timer)
reset_btn.grid(column=2, row=2)
changeOnHover(reset_btn, GREEN, YELLOW)

check_mark = Label(fg=GREEN, bg=YELLOW)
check_mark.grid(column=1, row=3)

window.mainloop( )
```
